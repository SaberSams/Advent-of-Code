```python
def parse(data):
  path = data.splitlines()[0]
  dict = {}
  for line in data.splitlines()[2:]:
    line = line.replace('(', '')
    line = line.replace(')', '')
    key, rest = line.split(' = ')
    dict[key] = {}
    dict[key]['L'], dict[key]['R'] = rest.split(', ')
  return path, dict

path, paths = parse(input)
print(paths)

steps = 0
current, end = 'AAA', 'ZZZ'

while current != end:
  next = paths[current][path[steps % len(path)]]
  current = next
  steps += 1
  if(steps % 1000000 == 0):
    print(steps)

print(steps)
```
